1- **DESCRIPCIÓN DEL PROBLEMA**  :

**Problema:** Encontrar un sub-arreglo dentro de una matriz que la suma de los números sea la mayor.


**Entrada**: Arreglo de n números. Ej [a1,a2,..an]

**Salida:** Sub-arreglo que la suma de sus elementos sea la mayor o igual a otros sub-arreglos.

**n** = Tamaño del problema.

**Aplicaciones:** Analisis de estadisticas graficas de numeros (por ejemplo: cantidad de automoviles en las carreteras )

2- **ALGORITMO A UTILIZAR**:

**SUB_ARREGLO MAX**

El algoritmo sub_arreglo máximo es el método para encontrar el subarray contiguo dentro de una matriz unidimensional de números que tiene la suma más grande.

*ENTRADA*: Un arreglo con n números.

*SALIDA*: La suma maxima de un sub-arreglo.


**Funcionamiento:**

EL algoritmo resive como entrada un arreglo de n elementos.


1.   Se divide el arreglo en sub-arreglos ( pueden ser de uno o más elementos consecutivos)
2.   Se calculan las sumas maximas de los sub-arreglos, primero parte izquierda y luego en la parte derecha (recursivamente)
3.   Se retorna la suma del sub-arreglo que sea la más grande.

**3- IMPLEMENTACIÓN DEL ALGORITMO**

Función **MaxSubArrayMid**

```python
def max_sub_array_mid(a, menor, media, mayor):

    iz = -1
    der = -1
    sum = 0
    
    sum_iz = float("-Inf")

    for i in range(media, menor - 1, -1):
        sum += a[i]
        if (sum > sum_iz):
            sum_iz = sum
            max_iz = i

    sum_der = float("-Inf")
    sum = 0

    for j in range(media + 1, mayor + 1):
        sum += a[j]
        if (sum > sum_der):
            sum_der = sum
            max_der = j
 
    return max_iz, max_der, sum_iz + sum_der
```

Función **MaxSubArray**

```python
import sys
 
def max_sub_array(n, iz=None, der=None, verbose = False):
 
    if not n:
        return 0

    if iz is None and der is None:
        iz, der = 0, len(n) - 1
 
    if der == iz:
        return n[iz]
 
    media = (iz + der) // 2
    max_iz = -sys.maxsize
    total = 0

    if verbose == True:
      print("Sub arreglo:",end="")

    for i in range(media, iz - 1, -1):
        total += n[i]
        if total > max_iz:
            max_iz = total
        if verbose == True:
          print("",n[i],end="")    

    max_der = -sys.maxsize
    total = 0  

    if verbose == True:
      print("")     
      print("Sub arreglo:",end="")
 
    for i in range(media + 1, der + 1):
        total += n[i]
        if total > max_der:
            max_der = total
        if verbose == True:
          print("",n[i],end="")

    if verbose == True:
      print("")
      max_der_iz = max(max_sub_array(n, iz, media,True),max_sub_array(n, media + 1, der,True))
      print()
      print(f"suma: {max_iz} + {max_der} : {max_iz + max_der}")
      print("")

    max_der_iz = max(max_sub_array(n, iz, media),max_sub_array(n, media + 1, der))

    if len(n) > 1:
      print("Índice inicial: ",max_iz)
      print("Índice final: ", max_der)
      print("")

    return max(max_der_iz, max_iz + max_der)
 
```

```python
n = [ -2, 1, -3, 4, -1, 2, 1, -5, 4 ]
print("ENTRADA: ",n)
suma_max = max_sub_array(n)
print('La suma máxima del sub_arreglo es:', suma_max)
```

**4- PROPIEDADES DEL ALGORITMO**


**CORRECTITUD**

***MaxSubArrayMid:***

- *INICIALIZACIÓN*

**Al tener un arreglo de un solo elemento**

Igualamos a valor 0 :

***suma_izquierda = 0***

***sum_derecha = 0***

***actual_suma_iz = 0***

***actual_suma_der = 0***

Luego se entra a un ***for*** para entrar al largo del arreglo, como solo tenemos un elemento, solo se recorre una sola vez y entra al ***if*** sin problema y luego se sale.

Para despues entrar nuevamente a otro ***for*** y luego al ***if*** con las mismas condiciones, por ello entra sin problemas.

- *MANTENCIÓN*

**Si el arreglo tiene dos elementos**

Al entrar al primer ***for***, se ejecutara dos veces y al cumplir los ***if** del codigo, se va a ejecutar de forma correcta recorriendo el ***for*** dos veces.

Entonces, podemos concluir que si al funcionar el codigo para un arreglo de dos elementos, es correcto para un arreglo de n elementos.


***MaxSubArray (*Por inducción*):***

- *INICIALIZACIÓN*

**Al tener un arreglo de largo n = 1**

Se debe verificar que la parte izquierda (índice inicial) sea igual a la parte derecha (índice final). 

*izquierda = 0*

*derecha = 0*

 ***a[izquierda] == a[derecha]***

y se retorna el arreglo con la posición inicial:  a[izquierda]

- *MANTENCIÓN*

**Si el arreglo es de largo n = 2**

Al ser diferentes:

 ***a[izquierda] != a[derecha]***

 Se divide en dos sub-arreglos:
 
 
 ***media = ( inicial + final) / 2***

y luego se llama de forma recursiva la función maxSubArray para la parte izquierda, la derecha y la parte del medio, retornando el máximo izquierdo, el máximo derecho y el máximo del medio.

Entonces podemos decir, que lo anterior es correcto para cualquier arreglo de tamaño n, ya que sirve para un arreglo de tamaño 2.

**5- EXPERIMENTOS Y ANÁLISIS**

Algoritmo de Kadane

```python
def kadane(A):
    max_so_far = 0
    max_ending_here = 0

    for i in A:
        max_ending_here = max_ending_here + i
        max_ending_here = max(max_ending_here, 0)
        max_so_far = max(max_so_far, max_ending_here)
 
    return max_so_far

a = [-2, 1, -3, 4, -1, 2, 1, -5, 4]
 
print("La suma máxima del sub-arreglo: ", kadane(a))
```

**Comparación algortimo Kanade y MaxSubArray**

```python
import matplotlib.pyplot as plt
import datetime
from timeit import repeat
import random

x=[]; y=[]; z=[]

for n in range(5,500):

  a = random.sample(range(1, 1000), n)
  t = repeat(setup="from __main__ import max_sub_array", stmt=f"max_sub_array({a})", repeat=1, number=10)
  e = repeat(setup="from __main__ import kadane", stmt=f"kadane({a})", repeat=1, number=10)
  x.append(n)
  y.append(t)
  z.append(e)


plt.plot(x,y,z)
plt.legend(["max_sub_array","kadane"])

plt.xlabel('n')
plt.ylabel('Tiempo en ms')
plt.show()
```

**ANÁLISIS DE RESULTADOS**

Como podemos observar, el tiempo de ejecución del kadane es menor al del maxSubArray, esto ocurre, ya que el algoritmo kadane es más eficiente para determinar la suma máxima, además de ser la mejora del algortimo maxSubArray.

El algortimo kadane tiene una complejidad de tiempo O(n), puesto que al recorrer el arreglo, cada pasada se actualiza el valor de la suma.

**Comparación de tiempo de ejecución de la función maxSubArray recursiva y no recursiva**

Función no recursiva

```python
def max_sub_array_noRecursivo(nums):
 
    maxSum = -sys.maxsize

    for i in range(len(nums)):
        total = 0  
        for j in range(i, len(nums)):
            total += nums[j]
            if total > maxSum:
                maxSum = total
    return maxSum
 
nums = [2, -4, 1, 9, -6, 7, -3]
print('La suma máxima es:', max_sub_array_noRecursivo(nums))
```

**Comparación de tiempo de ejecución entre el MaxSubArray recursivo y el no recursivo**

```python
import matplotlib.pyplot as plt
import datetime
from timeit import repeat
import random

x=[]; y=[]; z=[]

for n in range(5,500):

  a = random.sample(range(1, 1000), n)
  t = repeat(setup="from __main__ import max_sub_array", stmt=f"max_sub_array({a})", repeat=1, number=10)
  e = repeat(setup="from __main__ import max_sub_array_noRecursivo", stmt=f"max_sub_array_noRecursivo({a})", repeat=1, number=10)
  x.append(n)
  y.append(t)
  z.append(e)


plt.plot(x,y,z)
plt.legend(["MaxSubArray recursivo","MaxSubArray no recursivo"])

plt.xlabel('n')
plt.ylabel('Tiempo en ms')
plt.show()
```

**ANÁLISIS DE RESULTADOS**

Al comparar estos dos algoritmos, claramente se puede concluir que la forma recursiva es mejor que la no recursiva.

La no recursiva considera todos los arreglos posibles para encontrar la suma máxima, en cambio, la recursiva utiliza la técnica de Divide y Vencerás, dividiendo el arreglo en dos sub-arreglos iguales, calculando recursivamente la suma máxima para los sub-arreglos de la parte izquierda,derecha y la del medio, devolviendo la mayor suma.
