1- **DESCRIPCIÓN DEL PROBLEMA**  :

Problema: Ordenamiento de una secuencia de números.


Entrada: Secuencia de n números. Ej [a1,a2,..an]

Salida: Secuencia ordenada de los números de entrada de menor a mayor.

n = Tamaño del problema.



A.jpg

2- **ALGORITMO A UTILIZAR**:

**Insertion Sort**

Un algoritmo de ordenamiento que se asemeja a la idea de tener una baraja de cartas y querer ordenarlas de menor a mayor.
Es un metodo simple, facil de usar y eficiente si la cantidad de números a ordenar son pocos.

*ENTRADA*: En arreglo con n números.

*SALIDA*: El mismo arreglo de n números, pero ordenados de mayor a menor.


**Funcionamiento:**

El algoritmo resive como entrada un arreglo o lista con una secuencia de n números.

1.   Se toma el primer elemento ( a[0] )  del arreglo o lista. Este se dice que ya esta ordenado, por ello pasamos al siguiente elemento y se compara con el primero.
2.   Si el elemento con el que se esta comparando ***es mayor*** al número que se quiere ordenar, se intercambian de posición.
3.   Si ***es menor***, se quedan en la misma posición.





**EJEMPLO**

Algorithms_Insertion_Sort_Example.gif

Consideremos el arreglo que se muestra en la imagen:

  **a = [4,3,6,5]**



1.   Nos posicionamos en el número 4 (primera posición) y queda como el primer número ordenado, luego se pasa al número 3 y lo comparamos con el número 4, como el 4 es mayor a 3 se intercambian de posición.

  **a = [3,4,6,5]**

2.   Luego, el número 6 se compara con el 4, como el 4 es menor, no cambian de posición.

  **a = [3,4,6,5]**

3.   Por consiguiente, el número 5 se compara con el numero 6, como el 6 es mayor, intercambian posiciones.

  **a = [3,4,5,6]**


  Como ya no quedan más números, se termina el ciclo de comparaciones y el arreglo queda ordenado.

3- **IMPLEMENTACIÓN DEL ALGORITMO**

```python
import random
from termcolor import colored

def insertionSort(a, verbose = False ):

  aux = 0;
  n = len(a)
  if verbose == True: print("Arreglo:",a)

  for i in range(1, len(a)):
    if verbose == True: print("\nPasada",i)
    already_sorted = True
    clave = a[i]
    j = i - 1    

    while j >= 0 and clave < a[j]:
       aux = aux + 1
       a[j + 1] = a[j]
       if verbose == True: 
        print(str(a[:j])[1:-1],",",colored(f"{str(a[j:j+2])[1:-1]}","red"),",",str(a[j+2:])[1:-1])
       j = j - 1
       already_sorted = False
    a[j + 1] = clave

    if already_sorted == False and verbose == True: 
      print("\nAl finalizar pasada:")
      print(str(a[:n-i])[1:-1],",",colored(f"{str(a[n-i:])[1:-1]}","blue"))
    if already_sorted:
      if verbose == True: print("Esta ordenado!")
      
    
  if verbose == True: print("\noutput array:",a)


  return a, aux

data = [4,3,6,5]
print("ENTRADA:", data)
data, cont = insertionSort(data)
print("SALIDA:", data)
print("Las comparaciones son:",cont)

```

**4- PROPIEDADES DEL ALGORITMO**

**TIEMPO DE EJECUCCIÓN**

*TEOREMA*

El algoritmo de ordenamiento *INSERTION SORT* en el peor de los casos tiene una complejidad temporal de O(n^2) y en el mejor de los casos es O(n).


*PRUEBA DEL TEOREMA*

Para el peor de los casos, ocurre cuando el menor número del arreglo se encuentra en la ultima posición. 

Para el mejor de los casos, ocurre cuando los numeros del arreglo estan ordenados de forma ascendente. 
(En este caso, el algoritmo nunca ejecuta el ciclo *While*, ya que nunca se cumple la condición.)



**EJEMPLO**

PEOR CASO

```python
a = [7,5,3,1]
print("Entrada: ",a)
a, cont = insertionSort(a,verbose = True)
print("Salida: ",a)
print("Comparaciones: ",cont)
```

En este caso, en total fueron 6 comparaciones, al tener que mover el numero 1  (el menor elemento) a la primera posición.

MEJOR CASO

```python
a = [3,5,7,9]
print("Entrada: ",a)
a, cont = insertionSort(a,verbose = True)
print("Salida: ",a)
print("Comparaciones: ",cont)
```

Como se puede esperar, no tuvimos ninguna comparación, ya que el arreglo al estar ya ordenado no entro al ciclo while.

-------

**CORRECTITUD**

*TEOREMA*

Se genera un arreglo de entrada [a1,a2..an] y se espera que el algoritmo lo ordene de menor a mayor [a1 <= a2 <=... an].

*PRUEBA DEL TEOREMA*

(propiedad de bucle)

Como ya sabemos, *Insertion Sort* en cada interación el arreglo contiene los mismos elementos que el arreglo original ordenados de menor a mayor.

*INICIALIZACIÓN*

Sabemos que *i* tiene valor 1, entonces el primer elemento es a[i - 1], además sabemos que es el primer elemento que se encuentra ordenado.

*MANTENCIÓN*

Como sabemos, nuestro elemento a[i-1] esta ordenado. Al recorrer el arreglo el algoritmo toma el siguiente elemento y lo mueve a la posicion i que le corresponde. Como salida obtendremos el arreglo original ordenado.

LAS AFIRMACIONES CUMPLEN CON LAS PROPIEDADES DEL BUCLE.

**5- EXPERIMENTOS Y ANÁLISIS**

Número de comparaciones

MEJOR Y PEOR CASO

```python
import matplotlib.pyplot as plt

x=[n for n in range(5,20)] 
y1=[n*(n-1)/2 for n in range(5,20)] #peor  
y2=[n-1 for n in range(5,20)] #mejor
y=[]; 

for n in range(5,20):
  a = random.sample(range(1, 100), n)
  a,cont = insertionSort(a)
  y.append(cont)

plt.plot(x,y)
plt.plot(x,y1)
plt.plot(x,y2)
plt.legend(["Insertion Sort", "Peor caso", "Mejor caso"])

plt.xlabel('n')
plt.ylabel('número de operaciones')
plt.show()
```

**ANALISIS DE RESULTADOS**

Como podemos observar, las lineas de los tres casos estan bastante aisladas, esto significa que el algoritmo Insertion Sort, aunque sea efectivo para ordenar arreglos con pocos numeros, no es muy efectivo en general si los numeros estan muy desordenados o el tamaño del arreglo es muy grande.

Se puede identificar a simple vista que el algoritmo no tiene una preferencia hacia donde va.

UTILIZANDO ARREGLO CASI ORDENADO

```python
import matplotlib.pyplot as plt

x=[n for n in range(5,20)] 
y1=[n*(n-1)/2 for n in range(5,20)] #peor  
y2=[n-1 for n in range(5,20)] #mejor
y=[]; 

for n in range(5,20):
  a = [2,1,3,4,5,6]
  a,cont = insertionSort(a)
  y.append(cont)

plt.plot(x,y)
plt.plot(x,y1)
plt.plot(x,y2)
plt.legend(["Insertion Sort", "Peor caso", "Mejor caso"])

plt.xlabel('n')
plt.ylabel('número de operaciones')
plt.show()
```

Al querer utilizar un arreglo casi completamente ordenado, el algoritmo Insertion Sort se acerca al mejor de los casos.

**Comparacion Buble Sort e Insertion Sort**

Codigo bubble sort

```python
import random
def bubbleSort(a):
  n = len(a)
  t = 0
  for i in range ( 1, n-1):
    for j in range (n-i):
      t = t+1
      if a[j] > a[j+1]:
        a[j], a[j+1] = a[j+1], a[j]
  return a,t
arreglo = [3,6,7,2]
print("Entrada:",arreglo)
arreglo,cont = bubbleSort(arreglo)
print("Salida:",arreglo)
print("Comparaciones:",cont)
```

Comparación de tiempos de ejecución

```python
import matplotlib.pyplot as plt
import datetime
from timeit import repeat

x=[]; y=[]; z=[]

for n in range(5,500):

  a = random.sample(range(1, 1000), n)
  t = repeat(setup="from __main__ import insertionSort", stmt=f"insertionSort({a})", repeat=1, number=10)
  e = repeat(setup="from __main__ import bubbleSort", stmt=f"bubbleSort({a})", repeat=1, number=10)
  x.append(n)
  y.append(t)
  z.append(e)


plt.plot(x,y,z)
plt.legend(["Insertion Sort","Bubble Sort"])

plt.xlabel('n')
plt.ylabel('Tiempo en ms')
plt.show()
```

**ANALISIS DE RESULTADOS**

Como podemos ver en el grafico, el tiempo de ejecución de ambos algoritmos se comportan de manera similar, pero en 
diferentes tiempos.

El algoritmo Insertion Sort se demora menos en la ejecución, ya que no hace muchas comparaciones, en cambio, el algoritmo Bubble Sort se demora mucho más y más aun si el tamaño aumenta, ya que realiza más comparaciones que el Insertion Sort.
