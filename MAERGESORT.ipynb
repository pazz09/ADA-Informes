1- **DESCRIPCIÓN DEL PROBLEMA**  :

Problema: Ordenamiento de una secuencia de números.


Entrada: Secuencia de n números. Ej [a1,a2,..an]

Salida: Secuencia ordenada de los números de entrada de menor a mayor.

n = Tamaño del problema.






----------------------------------------------------------


2- **ALGORITMO A UTILIZAR**:

**Merge Sort**

Merge Sort es un algoritmo de ordenamiento de clasificación que se basa en el principio del "DIVIDE Y VENCERAS", además este se divide en dos funciones, merge y merge sort, este último resuelve el problema de forma recursiva.

*ENTRADA*: Un arreglo con n números.

*SALIDA*: El mismo arreglo de n números, pero ordenados de mayor a menor.


**Funcionamiento:**

En este algoritmo, el arreglo se divide inicialmente en dos mitades iguales y luego se combinan de manera ordenada.

1.   Dividir: Se divide el tamaño n del arreglo en 2. ( n/2 )
2.   Ordenar: Se ordenan los subarreglos de forma recursiva. ( merge sort )
3.   Mezclar: Se combinan los subarreglos de forma ordenada. ( merge )

**3- IMPLEMENTACION DEL ALGORITMO ( MERGE Y MERGE SORT )**

FUNCIÓN MERGE

```python
def merge(a1,a2):

  global aux

  a_ordenado = []
  i = 0
  j = 0

  while( i < len(a1) and  j < len(a2)):
    
    if (a1[i] < a2[j]):
      a_ordenado.append(a1[i])
      i = i + 1
    else:
      a_ordenado.append(a2[j])
      j = j + 1
    aux = aux + 1

  a_ordenado = a_ordenado + a1[i:]
  a_ordenado = a_ordenado + a2[j:]


  return a_ordenado
```

FUNCION MERGE SORT

```python
def mergeSort(a, verbose = False):
  global cont
  global n

  if verbose == True:
    print("Entrada",n,":",a)
    n = n + 1

  if len(a) < 2:
    return a
  else:
    x = len(a) // 2

    if verbose == True:
      y = mergeSort(a[:x], verbose = True)
      z = mergeSort(a[x:], verbose = True)

    if verbose == False:
      y = mergeSort(a[:x])
      z = mergeSort(a[x:])
    cont = cont + 1
    
    if verbose == True :
      print("Arreglo ordenado: ", merge(y,z))
      print("\n")

    return merge(y,z)
```

**4- PROPIEDADES DEL ALGORITMO**

**CORRECTITUD** 

TEOREMA

Se genera un arreglo de entrada [a1,a2..an] y se espera que el algoritmo lo ordene de menor a mayor [a1 <= a2 <=... an].

PRUEBA DEL TEOREMA

Para probar el teorema se necesitara un análisis matemático.

Declaracion: P(n)

Cantidad de números : n

- Merge Sort, retorna un arreglo conn **n** elementos de entrada ordenados de menor a mayor. 

- Si n = 1 , Merge Sort retorna el arreglo de 1 elemento sin hacer ninguna comparación.

- Un arreglo de **n** elementos, se divide en 2 sub-arreglos de tamaño n/2 y se ordenan usando la funcion **merge sort**. Se sabe que estan ordenados, ya que el tamaño de los sub-arreglos es menor al tamaño del arreglo original. 
Por ultimo, la función merge retorna el arreglo ordenado mezclando los subarreglos.

Con lo dicho, podemos probar que son correctas las siguientes afirmaciones:



1.   P(n) para un caso base, es P(1).
2.   P(m) es verdadero para m < n, entonces P(n) también lo es.



**TIEMPO DE EJECUCCIÓN**

TEOREMA

El algoritmo MERGE SORT es una función recursiva, por ello, para explicar el tiempo de ejecución se puede utilizar una ecuación matemática recursiva.

ANALISIS

Tamaño del problema: N

- Si n es muy pequeño, el tiempo que requerirá el algoritmo será invariante, es decir: **T(N) = O(1)**

- Si n esta en el promedio, se divide el problema en **a** veces sub-arreglos y cada uno es **1/b** del arreglo original. El algoritmo se demorará **a * T(N/b)**  en ordenar los sub-arreglos.

- Por ultimo el algoritmo se demora **D(N)** en dividir el arreglo en **a** sub-arreglos y **C(N)** en combinarlos.

**- Comparacion del algoritmo MergeSort con el InsertionSort**

Codigo insertion sort

```python
def insertionSort(a, verbose = False ):
  aux = 0;
  n = len(a)
  for i in range(1, len(a)):
    clave = a[i]
    j = i - 1    
    while j >= 0 and clave < a[j]:
       aux = aux + 1
       a[j + 1] = a[j]
       j = j - 1
    a[j + 1] = clave
  return a, aux

data = [4,3,6,5]
print("ENTRADA:", data)
data, cont = insertionSort(data)
print("SALIDA:", data)
print("Las comparaciones son:",cont)
```
