1- **DESCRIPCIÓN DEL PROBLEMA**  :

Problema: Ordenamiento de una secuencia de números.


Entrada: Secuencia de n números. Ej [a1,a2,..an]

Salida: Secuencia ordenada de los números de entrada de menor a mayor.

n = Tamaño del problema.






----------------------------------------------------------


2- **ALGORITMO A UTILIZAR**:

**Quick Sort**

Quicksort es un algoritmo de ordenamiento rápido, basado en la idea de divide y vencerás.




*ENTRADA*: Un arreglo con n números.

*SALIDA*: El mismo arreglo de n números, pero ordenados de menor a mayor.


**Funcionamiento:**

1.  El arreglo se divide en sub-arreglos y se elige un elemento pivote (elemento seleccionado del arreglo).

2.  Al dividir el arreglo, el elemento pivote debe colocarse de tal manera que los elementos menores que el pivote se mantengan en el lado izquierdo y los elementos mayores que el pivote estén en el lado derecho del pivote.

3.  Los sub-arreglos izquierdo y derecho también se dividen. Este proceso continúa hasta que cada sub-arreglo contiene un solo elemento.

4.  En este punto, los elementos ya están ordenados. Por último, los elementos se combinan para formar un arreglo ordenado.




**EJEMPLO**

ejemplo.jpeg

Consideremos el arreglo:

a = [5, 2, 7, 3, 9, 4, 8]

1. Elegimos como pivote el primer elemento:

***pivote = a[0]***

***a[0] = 5***

2. Buscamos un elemento "x" que sea mayor que el pivote, de izquierda a derecha.


3. Buscamos un elemento "y" que sea menor que el pivote de derecha a izquierda.

4. Si el elemento x < y , se intercambian los valores.

5. Se repide el procedimiento.

3- **IMPLEMENTACIÓN DEL ALGORITMO**

FUNCIÓN PARTITION

```python
def partition(array, low, high,verbose = False ):

  if verbose == True:
    print("")
    print("Arreglo inicial:",array)

  
  pivot = array[high]

  if verbose == True:
    print("Pivote:",pivot)

  i = low - 1

  for j in range(low, high):
    
    if array[j] <= pivot:
      i = i + 1
      (array[i], array[j]) = (array[j], array[i])

  (array[i + 1], array[high]) = (array[high], array[i + 1])

  if verbose == True:
    print("Arreglo final:", array)
    print("")

  return i + 1
```

FUNCIÓN QUICK SORT

```python
def quick_Sort(arr, bajo, alto, verbose = False):

  global cont

  if bajo < alto:

    cont = cont + 1

    if verbose == True:
      pi = partition(arr, bajo, alto, verbose = True)
      quick_Sort(arr, bajo, pi - 1,verbose = True)
      quick_Sort(arr, pi + 1, alto, verbose = True)
      
    else:
      pi = partition(arr, bajo, alto)
      quick_Sort(arr, bajo, pi - 1)
      quick_Sort(arr, pi + 1, alto)

    return arr

```

RESULTADOS:

Usando como pivote el último elemento

(La función verbose esta desactivada)

```python
a = [5, 2, 7, 3, 9, 4, 8]

print("ENTRADA: ", a)

size = len(a)
cont = 0

a_ordenado = quick_Sort(a, 0, size - 1)

print("SALIDA: ", a_ordenado)
print("Iteraciones: ",cont)
```

**4- PROPIEDADES DEL ALGORITMO**

**CORRECTITUD**

Función **PARTITION**

-----------------

- *TEOREMA*

La función partition recibe un arreglo a[BAJO.....ALTO], lo organiza y retorna un indicador. Este indicador divide el arreglo en dos, los elementos menores o iguales al indicador y los elementos del arreglo que son mayores al indicador.

**PROPIEDAD INVARIANTE DEL BUCLE**

Al iniciar las comparaciones, se cumple que: Para cualquier indice:

1. Si BAJO <= INDICE <= i (puntero)
Entonces:
**a[INDICE] <= a[ALTO]**
2. Si i <= INDICE < j(indice que recorre)
Entonces:
**a[INDICE] > a[ALTO]**

-----------------
- *PRUEBA DEL TEOREMA*

INICIALIZACIÓN

1. Al comienzo de la primera iteración:

i = BAJO - 1 

j = BAJO

2. Como el INDICE no se encuentra en la primera iteración:

BAJO <= INDICE <= i

o

i < INDICE < j

**Por ello, las propiedades invariantes son ciertas.**


-----------------------------------------
**TIEMPO DE EJECUCIÓN**

- *TEOREMA*

Funcion de recurrencia: 


**T(n)= 2T(n/2)+n**

a=2 , b=2, d=1

Por teorema maestro: a = b^d

Caso 1:  O( n log n )

---------------

- *PRUEBA DEL TEOREMA**

El peor caso ocurre cuando el pivote reiteradas veces se inclina hacia un lado en especifico, es decir, siempre queda un arreglo de 1 elemento. 

- Tiempo de ejecución:

PEOR CASO: O(n^2)
MEJOR CASO (Caso promedio): O( n log n)

COMPARACIÓN PEOR Y MEJOR CASO

```python
import matplotlib.pyplot as plt
import random

x=[n for n in range(5,20)] 
y1=[n*(n-1)/2 for n in range(5,20)] #peor  
y2=[n-1 for n in range(5,20)] #mejor
y=[]
cont = 0

for n in range(5,20):
  b = random.sample(range(1, 100), n)
  size = len(b)
  b = quick_Sort(b, 0, size - 1)
  y.append(cont)

plt.plot(x,y)
plt.plot(x,y1)
plt.plot(x,y2)
plt.legend(["Quick Sort", "Peor caso", "Mejor caso"])

plt.xlabel('n')
plt.ylabel('número de operaciones')
plt.show()
```

**ANÁLISIS DE RESULTADOS**

Primero que nada, en este experimento se utilizó como pivote el último elemento.

El grafico nos señala que el algoritmo *quick sort* se inclina más a su peor caso. Esto ocurre, ya que, el algoritmo, solo es eficiente con pocos elementos, cuando el tamaño del arreglo es pequeño.

COMPARACIÓN DE TIEMPOS DE EJECUCIÓN ENTRE EL **MERGE SORT** Y EL **QUICK SORT**

```python
def merge(a1,a2):
  a_ordenado = []
  i = 0
  j = 0
  while( i < len(a1) and  j < len(a2)):
    if (a1[i] < a2[j]):
      a_ordenado.append(a1[i])
      i = i + 1
    else:
      a_ordenado.append(a2[j])
      j = j + 1
  a_ordenado = a_ordenado + a1[i:]
  a_ordenado = a_ordenado + a2[j:]
  return a_ordenado
```

```python
def mergeSort(a, verbose = False):

  if len(a) < 2:
    return a
  else:
    x = len(a) // 2
    y = mergeSort(a[:x])
    z = mergeSort(a[x:])
    
    return merge(y,z)
```

```python
import matplotlib.pyplot as plt
import datetime
from timeit import repeat

x=[]; y=[]; z=[]
cont = 0

for n in range(5,500):

  a = random.sample(range(1, 1000), n)
  size = len(a)
  t = repeat(setup="from __main__ import quick_Sort", stmt=f"quick_Sort({a},{0}, {size - 1})", repeat=1, number=10)
  e = repeat(setup="from __main__ import mergeSort", stmt=f"mergeSort({a})", repeat=1, number=10)
  x.append(n)
  y.append(t)
  z.append(e)


plt.plot(x,y,z)
plt.legend(["Quick Sort","Merge Sort"])

plt.xlabel('n')
plt.ylabel('Tiempo en ms')
plt.show()
```

**ANÁLISIS DE RESULTADOS**

Como podemos ver en el grafico, el algoritmo Merge Sort tiene un mayor tiempo de ejecucción que el Quick Sort a medida que el tamaño del arreglo aumenta.

Esta diferencia, ocurre por la forma de como ordenan los elementos del arreglo. El quick sort ordena los elementos comparando cada uno con un pivote, en cambio, el merge sort divide el arreglo en dos sub-arreglos una y otra vez hasta que queda solo un elemento.

Por ello la velocidad de la ejecuccion del quick sort es más rápida.
